```python
import csv
import networkx as nx
import scipy as np
```

```python
### Extracting nodes from csv
with open('bis_vertices-1.csv', 'r') as nodecsv: # Open the file
    nodereader = csv.reader(nodecsv) # Read the csv
    nodes = [n for n in nodereader][1:] # Retrieve the data
    
nodes_2106 = [n for n in nodes if n[0]=="2021-06-30"]
node_names_2106 = [n[1] for n in nodes_2106] # Get a list of only the node names
node_names_2106.remove("5A")
print(len(node_names_2106))
#print(node_names_2106)
```

```python
### Extracting the edges from csv
with open('bis_arcs-1.csv', 'r') as edgecsv: #Open the file
    edgereader = csv.reader(edgecsv) # Read the csv
    edges = [e for e in edgereader][1:] # Retrieve the data
    
edges_2106 = [tuple(e[1:4]) for e in edges if e[0]=="2021-06-30"]
edges_2106 = [e for e in edges_2106 if "5A" not in e]
#print(len(edges_2106))
#print(edges_2106)
```

```python
### Creating the network
G = nx.DiGraph()

G.add_nodes_from(node_names_2106)
G.add_weighted_edges_from(edges_2106)

#print(nx.info(G))
#for (u, v, wt) in G.edges.data('weight'):
#    print(f"({u}, {v}, {wt:.7})")
```

```python
### Examining elements of the network
list(G.nodes)
list(G.edges)
list(G.adj["US"])
G.degree["US"]
G.edges(["US", "IN"])
G.degree(["UK", "GB", "TR"])
```

```python
### Analysing the network
nx.density(G)
nx.average_clustering(G)
nx.is_strongly_connected(G)

### Need to separete the analysis for different connected components
#nx.periphery(G) 
#nx.diameter(G)
```

```python
### Analysing the nodes
nx.clustering(G)
nx.degree_centrality(G)
nx.average_neighbor_degree(G)
nx.eigenvector_centrality(G)
b = dict(nx.betweenness_centrality(G))
degree = nx.degree(G)
d = dict(G.degree)
#nx.pagerank(G)
```

```python
### Analysing the edges
nx.edge_betweenness_centrality(G)
```

```python
### Detecting communities/clusters in network
from networkx.algorithms.community import greedy_modularity_communities
c = list(greedy_modularity_communities(G))
sorted(map(sorted, c))
```

```python
from networkx.algorithms import community
communities_generator = community.girvan_newman(G)
top_level_communities = next(communities_generator)
next_level_communities = next(communities_generator)
sorted(map(sorted, next_level_communities))
```

```python
### Network Visualisation
import matplotlib.pyplot as plt
nx.draw(G)
```

```python
weight = [e[2] for e in edges_2106]
nx.draw(G, pos=nx.circular_layout(G), node_color='g', edge_color='k', 
        node_size=[v * 250000 for v in b.values()], alpha=0.5, 
        width=[float(w)/100000 for w in weight])
labels = nx.draw_networkx_labels(G, pos=nx.circular_layout(G))
```
